```python
# -*- coding: utf-8 -*-

from math import pow
from math import log2
from math import factorial

N=9

# benchmark data generation
brute = [1,8,64,512,4096,32768,262144,2097152,16777216]
Strassen = [0,7,56,399,2800,19607,137256,960799,6725600]
yVals1 = [] # cubic for O(n^3)
yVals2 = [] # slightly less than cubic for O(n^2.8)

# generate bench mark data for O(n^3), O(n^2.8)
for i in [1,2,4,8,16,32,64,128,256]:
    #brutecounts.append(round(pow(i, 2.99), 2))
    #Strassencounts.append(round(pow(i, 2.75), 2))
    yVals1.append(pow(i, 3))
    yVals2.append(round(pow(i, 2.8), 2))

# %matplotlib inline
import matplotlib.pyplot as plt

# Brute vs. Strassen execution steps
plt.rcParams['figure.figsize'] = [10,10]
fig, (ax1) = plt.subplots(1, 1)
fig.suptitle('Runtime Analysis', size="xx-large")

ax1.set_title('Brute Force vs. Strassen', size="x-large")
ax1.plot(range(N), brute, "rx-", markersize=10, linewidth=4, label="Brute")
ax1.plot(range(N), Strassen, "gx-", markersize=10, linewidth=4, label="Strassen")
ax1.plot(range(N), yVals1, "black", markersize=5, linewidth=2, label="benchmark O(n^3)")
ax1.plot(range(N), yVals2, "orange", markersize=5, linewidth=2, label="benchmark O(n^2.8)")
ax1.grid(b=True, which='major', axis='both')
ax1.set(xlabel = 'Matrix Order (2^n)', ylabel = 'Number of Multiplications')
ax1.legend(loc=(0.05,0.65), scatterpoints=1)
ax1.tick_params(axis="both", which="major", labelsize=14)
```
